```python
import tkinter as tk
from tkinter import messagebox
from getpass import getpass

# Sample data
books = [
    {"id": 1, "title": "Book 1", "author": "Author 1", "copies": 5},
    {"id": 2, "title": "Book 2", "author": "Author 2", "copies": 3},
    {"id": 3, "title": "Book 3", "author": "Author 3", "copies": 2},
]

users = [
    {"username": "admin", "password": "admin", "role": "admin"},
    {"username": "user1", "password": "pass1", "role": "user"},
    {"username": "user2", "password": "pass2", "role": "user"},
]

transactions = []

# Functions
def login(username, password):
    for user in users:
        if user["username"] == username and user["password"] == password:
            return user
    return None

def open_admin_window(user):
    admin_window = tk.Toplevel(root)
    admin_window.title("Admin Menu")

    # Admin menu content goes here
    label = tk.Label(admin_window, text=f"Welcome, {user['username']}", padx=20, pady=20)
    label.pack()

    button_view_transactions = tk.Button(admin_window, text="View Transactions", command=view_transactions)
    button_view_transactions.pack(pady=5)

    button_add_book = tk.Button(admin_window, text="Add Book", command=add_book)
    button_add_book.pack(pady=5)

    button_logout = tk.Button(admin_window, text="Logout", command=admin_window.destroy)
    button_logout.pack(pady=5)

def open_user_window(user):
    user_window = tk.Toplevel(root)
    user_window.title("User Menu")

    # User menu content goes here
    label = tk.Label(user_window, text=f"Welcome, {user['username']}", padx=20, pady=20)
    label.pack()

    button_search_book = tk.Button(user_window, text="Search Book", command=search_book_window)
    button_search_book.pack(pady=5)

    button_issue_book = tk.Button(user_window, text="Issue Book", command=issue_book_window)
    button_issue_book.pack(pady=5)

    button_return_book = tk.Button(user_window, text="Return Book", command=return_book_window)
    button_return_book.pack(pady=5)

    button_logout = tk.Button(user_window, text="Logout", command=user_window.destroy)
    button_logout.pack(pady=5)

def add_book():
    add_book_window = tk.Toplevel(root)
    add_book_window.title("Add Book")

    label_title = tk.Label(add_book_window, text="Title:")
    label_title.grid(row=0, column=0, padx=5, pady=5)
    entry_title = tk.Entry(add_book_window)
    entry_title.grid(row=0, column=1, padx=5, pady=5)

    label_author = tk.Label(add_book_window, text="Author:")
    label_author.grid(row=1, column=0, padx=5, pady=5)
    entry_author = tk.Entry(add_book_window)
    entry_author.grid(row=1, column=1, padx=5, pady=5)

    label_copies = tk.Label(add_book_window, text="Copies:")
    label_copies.grid(row=2, column=0, padx=5, pady=5)
    entry_copies = tk.Entry(add_book_window)
    entry_copies.grid(row=2, column=1, padx=5, pady=5)

    def add_book_to_library():
        title = entry_title.get()
        author = entry_author.get()
        copies = entry_copies.get()
        if title and author and copies:
            books.append({"id": len(books) + 1, "title": title, "author": author, "copies": int(copies)})
            messagebox.showinfo("Success", "Book added successfully.")
            add_book_window.destroy()
        else:
            messagebox.showerror("Error", "Please fill all fields.")

    button_add = tk.Button(add_book_window, text="Add Book", command=add_book_to_library)
    button_add.grid(row=3, column=0, columnspan=2, padx=5, pady=5)

def view_transactions():
    transaction_list = "\n".join([f"{transaction['user']} {transaction['action']} {transaction['book']}" for transaction in transactions])
    messagebox.showinfo("Transaction History", "Transaction History:\n" + transaction_list)

def search_book_window():
    search_window = tk.Toplevel(root)
    search_window.title("Search Book")

    label_title = tk.Label(search_window, text="Enter Book Title:")
    label_title.pack(pady=5)
    entry_title = tk.Entry(search_window)
    entry_title.pack(pady=5)

    def search_book():
        title = entry_title.get()
        if title:
            result = search_book(title)
            if result:
                messagebox.showinfo("Book Found", f"Book found: {result['title']} by {result['author']}")
            else:
                messagebox.showinfo("Book Not Found", "Book not found.")

    button_search = tk.Button(search_window, text="Search", command=search_book)
    button_search.pack(pady=5)

def search_book(title):
    for book in books:
        if book["title"].lower() == title.lower():
            return book
    return None

def issue_book_window():
    issue_window = tk.Toplevel(root)
    issue_window.title("Issue Book")

    label_title = tk.Label(issue_window, text="Enter Book Title:")
    label_title.grid(row=0, column=0, padx=5, pady=5)
    entry_title = tk.Entry(issue_window)
    entry_title.grid(row=0, column=1, padx=5, pady=5)

    def issue_book():
        title = entry_title.get()
        if title:
            book = search_book(title)
            if book:
                if book["copies"] > 0:
                    book["copies"] -= 1
                    transactions.append({"user": current_user["username"], "action": "Issue", "book": book["title"]})
                    messagebox.showinfo("Success", f"Book '{book['title']}' issued successfully.")
                    issue_window.destroy()
                else:
                    messagebox.showerror("Error", "Sorry, the book is out of stock.")
            else:
                messagebox.showerror("Error", "Book not found.")
        else:
            messagebox.showerror("Error", "Please enter book title.")

    button_issue = tk.Button(issue_window, text="Issue", command=issue_book)
    button_issue.grid(row=1, column=0, columnspan=2, padx=5, pady=5)

def return_book_window():
    return_window = tk.Toplevel(root)
    return_window.title("Return Book")

    label_title = tk.Label(return_window, text="Enter Book Title:")
    label_title.grid(row=0, column=0, padx=5, pady=5)
    entry_title = tk.Entry(return_window)
    entry_title.grid(row=0, column=1, padx=5, pady=5)

    def return_book():
        title = entry_title.get()
        if title:
            book = search_book(title)
            if book:
                book["copies"] += 1
                transactions.append({"user": current_user["username"], "action": "Return", "book": book["title"]})
                messagebox.showinfo("Success", f"Book '{book['title']}' returned successfully.")
                return_window.destroy()
            else:
                messagebox.showerror("Error", "Book not found.")
        else:
            messagebox.showerror("Error", "Please enter book title.")

    button_return = tk.Button(return_window, text="Return", command=return_book)
    button_return.grid(row=1, column=0, columnspan=2, padx=5, pady=5)

def open_login_window():
    login_window = tk.Toplevel(root)
    login_window.title("Login")

    frame_login = tk.Frame(login_window)
    frame_login.pack(padx=20, pady=20)

    label_username = tk.Label(frame_login, text="Username:")
    label_username.grid(row=0, column=0, padx=5, pady=5)
    entry_username = tk.Entry(frame_login)
    entry_username.grid(row=0, column=1, padx=5, pady=5)

    label_password = tk.Label(frame_login, text="Password:")
    label_password.grid(row=1, column=0, padx=5, pady=5)
    entry_password = tk.Entry(frame_login, show="*")
    entry_password.grid(row=1, column=1, padx=5, pady=5)

    button_login = tk.Button(frame_login, text="Login", command=lambda: on_login_click(login_window, entry_username.get(), entry_password.get()))
    button_login.grid(row=2, column=0, columnspan=2, padx=5, pady=5, sticky="we")

def on_login_click(login_window, username, password):
    global current_user
    current_user = login(username, password)
    if current_user:
        messagebox.showinfo("Login Successful", f"Welcome, {current_user['username']}!")
        if current_user["role"] == "admin":
            open_admin_window(current_user)
        else:
            open_user_window(current_user)
        login_window.destroy()
    else:
        messagebox.showerror("Login Failed", "Invalid username or password.")

# GUI setup
def main():
    global root
    root = tk.Tk()
    root.title("Library Management System")

    label_welcome = tk.Label(root, text="Welcome to Library Management System", padx=20, pady=20)
    label_welcome.pack()

    button_admin_login = tk.Button(root, text="Admin Login", command=open_login_window)
    button_admin_login.pack(pady=5)

    button_user_login = tk.Button(root, text="User Login", command=open_login_window)
    button_user_login.pack(pady=5)

    root.mainloop()

if __name__ == "__main__":
    main()
```
